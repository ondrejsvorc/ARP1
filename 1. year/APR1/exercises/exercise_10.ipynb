```python
# Testování software
# - cílem je zjistit, zda software splňuje jak funkční, tak i nefunkční požadavky
# - Funkční požadavky = požadavky, které popisují, jaké funkce software musí mít, aby splnil potřeby zákazníka
# - Nefunkční požadavky = požadavky, které popisují vlastnosti softwaru
# - Verifikace (automatizované testování) - naprogramoval to programátor správně?
# - Validace (manuální testování) - dělá to, co chtěl klient?
# - Iterativní metodiky vývoje: SCRUM, Extreme Programming

# TDD (Test Driven Development)
# - vývoj řízený testy
# - AAA - Arrange, Act, Assert
# - Arrange = příprava dat
# - Act = vykonání nějaké akce s daty
# - Assert = ověření správnosti výsledku
# PRINCIP TDD:
# - nejprve se píšou testy k metodě, která bude vykonávat nějakou činnost
# - metoda se postupně píše tak, aby prošla napsanými testy
# - tím se odhalí, jak asi bude vypadat struktura metody
# - jakmile máme napsané testy, napíšeme funkcionalitu dané metody
# - při implementaci metody ji můžeme průběžně spolehlivě testovat, protože máme již napsané testy
# - jakmile ji budeme mít dopsanou, tak nejenom, že si můžeme být s velkou pravděpodobností jistí, že je napsaná správně,
#   ale ještě k tomu k ní máme napsané de facto všechny potřebné testy.
# - Kolik je potřeba testů? Tolik, kolik je domén ekvivalence
# - např. 3 + 4 a 2 + 3 = bude se chovat stejně (=je to stejná doména ekvivalence)
# - nezapomínat na extrémní hodnoty
# - PyTest

# Scrum:
# - vyvijí se v tzv. iteracích (sprintech)
# - Scum Master = komunikuje s Product Owner a s týmem vývojářů
# - Product Owner = píše Product backlog, stará se o to, aby projekt dopadl dobře
# - Product backlog = seznam všech požadavků na software
# - Sprint backlog = seznam požadavků, které se budou implementovat v aktuálním sprintu
# - Sprint = 1-4 týdny dlouhé období, během kterého se implementuje sprint backlog
# - Scrum meeting (daily stand up) = krátké (maximálně 15 minut) denní setkání, na kterém se tým dozvídá o pokroku ostatních členů týmu
# - Sprint review = setkání, na kterém se zákazníkovi předvádí výsledky aktuálního sprintu
# - Sprint retrospective = setkání, na kterém se tým zabývá zlepšeními pro následující sprint

# - Extreme programming:
# - Pair programming - Navigator, Driver - 2 programátoři u jednoho počítače
# - TDD
```

```python
from numpy import average, median


def get_statistics(file_path):
    try:
        with open(file_path, "r") as file:
            line = list(map(int, file.readline().split(" ")))
            avg = average(line)
            med = median(line)
            return (avg, med)
    except FileNotFoundError:
        return "Soubor neexistuje"
    except ValueError:
        return "Obsahuje pismenko"


def test_file_does_not_exist():
    assert get_statistics("fenjifsjisdfjisfd.txt") == "Soubor neexistuje"


def run_tests():
    test_file_does_not_exist()


def main():
    ...


run_tests()
main()
```

```python
def get_numbers_divisible_by(divisor: int, numbers: list):
    for number in numbers:
        if number % divisor == 0:
            yield number


divisor = 7
numbers = [7, 14, 21, 2, 3, 4]
result = list(get_numbers_divisible_by(divisor, numbers))

print(result)
```

```python
def get_numbers_divisible_by(divisor: int, numbers: list):
    return list(filter(lambda number: number % divisor == 0, numbers))
```

```python
import string
from typing import List


def get_alphabet_letters_at(indexes: List[int]) -> List[str]:
    for index in indexes:
        yield string.ascii_lowercase[index]


indexes = [0, 1, 2]
result = list(get_alphabet_letters_at(indexes))

print(result)
```

```python
def print_matrix(width: int, height: int) -> None:
    current_number = 0
    row = []
    while height > 0:
        for _ in range(0, width):
            row.append(current_number)
            current_number += 1
        print(row)
        row = []
        height -= 1


print_matrix(5, 5)
```

```python
from random import randint
from typing import Tuple


def create_matrix(
    width: int, height: int, treshold: Tuple[int, int]
) -> List[list[list]]:
    matrix = []
    for _ in range(height):
        row = []
        for _ in range(width):
            row.append(randint(treshold[0], treshold[1]))
        matrix.append(row)
    return matrix


matrix = create_matrix(width=10, height=10, treshold=(0, 255))
print(matrix)
```

```python
def are_anagrams(wordFirst: str, wordSecond: str) -> bool:
    return sorted(wordFirst) == sorted(wordSecond)


print(
    "Slova jsou anagramem."
    if are_anagrams("tes", "test")
    else "Slova nejsou anagramem."
)
```

```python
from collections import Counter


def are_anagrams(wordFirst: str, wordSecond: str) -> bool:
    return Counter(wordFirst) == Counter(wordSecond)


print(
    "Slova jsou anagramem."
    if are_anagrams("tes", "test")
    else "Slova nejsou anagramem."
)
```

```python
import matplotlib.pyplot as plt


def draw_matrix(matrix: List[list[int]]):
    plt.imshow(matrix, cmap="gray")


print(draw_matrix(create_matrix(100, 100, (0, 255))))
```
